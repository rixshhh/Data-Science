## Formula Explanation

### üìè Distance of a Point from a Plane

If the equation of a plane is:

Ax + By + Cz + D = 0


and the coordinates of a point are:

P(x_1, y_1, z_1)


then the **perpendicular distance (d)** of the point from the plane is given by:

d = |Ax_1 + By_1 + Cz_1 + D| / sqrt of (A^2 + B^2 + C^2)


---

#### üí° **Explanation**
- (A, B, C)  ‚Üí the normal vector to the plane  
- D ‚Üí the constant term  
- (x_1, y_1, z_1) ‚Üí the coordinates of the point  

This formula gives the **shortest distance** between the point and the plane.


## Example Calculation

```python
import math

# Given plane: 2x - 3y + 6z - 4 = 0
A, B, C, D = 2, -3, 6, -4

# Given point: (1, 2, 3)
x1, y1, z1 = 1, 2, 3

# Distance formula
distance = abs(A*x1 + B*y1 + C*z1 + D) / math.sqrt(A**2 + B**2 + C**2)
print(f"Distance of point ({x1}, {y1}, {z1}) from plane {A}x + {B}y + {C}z + {D} = 0 is {distance:.2f}")
```

## Visualization of Point and Plane

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Plane: 2x - 3y + 6z - 4 = 0
A, B, C, D = 2, -3, 6, -4

# Point (1, 2, 3)
x1, y1, z1 = 1, 2, 3

# Create grid for the plane
x = np.linspace(-2, 4, 10)
y = np.linspace(-2, 4, 10)
X, Y = np.meshgrid(x, y)
Z = (-A*X - B*Y - D) / C

# Plot the plane and point
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, alpha=0.5, color='cyan', label='Plane')
ax.scatter(x1, y1, z1, color='red', s=50, label='Point P(1,2,3)')

# Draw a line showing the perpendicular distance
x_line = np.linspace(x1, x1, 10)
y_line = np.linspace(y1, y1, 10)
z_line = np.linspace(z1, (-A*x1 - B*y1 - D)/C, 10)
ax.plot3D(x_line, y_line, z_line, 'k--', label='Perpendicular Distance')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Distance of a Point from a Plane')
plt.legend()
plt.show()
```

# ü§ñ Types of Machine Learning Techniques

Machine Learning (ML) techniques are broadly divided into **four main types**, based on how the model learns from data and the kind of feedback it receives.

## **1. Supervised Learning**
**Definition:**  
The model is trained using **labeled data** ‚Äî that means both input and the correct output are known. The algorithm learns to map inputs to outputs.

Output Feature will be continous -> Regression Problem Statement.

If it's not continous (Categorical) -> Then Classification Problem Statement.

* If there are two categories we basically say it as **Binary Classification**. 

* If there are multiple Categories then it becomes **MultiClass Classification**.

**Goal:**  
Predict outcomes for **new, unseen data**.

**Examples:**
- Predicting house prices based on area, rooms, etc.  
- Email spam detection (Spam / Not Spam)  
- Credit risk prediction (Low / High risk)  

**Common Algorithms:**  
- Linear Regression  
- Logistic Regression  
- Decision Trees  
- Random Forest  
- Support Vector Machines (SVM)  

üß© *Example:* If you give the model many examples of emails labeled ‚Äúspam‚Äù or ‚Äúnot spam,‚Äù it will learn to classify future emails automatically.

## **2. Unsupervised Learning**
**Definition:**  
In unsupervised learning, the data is **unlabeled** ‚Äî the model tries to find hidden patterns or structure in the data without knowing the correct output.

**Goal:**  
Group or organize data based on similarity or detect hidden patterns.

**Examples:**
- Customer segmentation in marketing  
- Grouping news articles by topic  
- Detecting abnormal network activity  

**Common Algorithms:**  
- K-Means Clustering  
- Hierarchical Clustering  
- Principal Component Analysis (PCA)  
- Association Rules (Apriori Algorithm) 
- DBSCAN 

üß© *Example:* A retail company can use clustering to divide customers into groups with similar buying habits.

## **3. Semi-Supervised Learning**
**Definition:**  
Semi-supervised learning uses **a mix of labeled and unlabeled data**. The model learns from the small labeled portion and generalizes to the large unlabeled portion.

**Goal:**  
Improve learning accuracy when labeled data is expensive or hard to get.

**Examples:**
- Web page classification  
- Medical image diagnosis (few labeled cases, many unlabeled)  
- Speech recognition  

**Common Algorithms:**  
- Self-training  
- Graph-based algorithms  
- Semi-supervised Support Vector Machines (S3VM)  

üß© *Example:* A healthcare model might have only a few labeled MRI scans but many unlabeled ones ‚Äî semi-supervised learning can still build an accurate model.

## **4. Reinforcement Learning**
**Definition:**  
Reinforcement Learning (RL) trains an agent to **make decisions by interacting with an environment**. The agent learns through **trial and error**, receiving **rewards or penalties** for actions.

**Goal:**  
Maximize cumulative rewards over time by learning the best strategy (policy).

**Examples:**
- Self-driving cars learning to drive safely  
- Game playing (Chess, Go, or Atari games)  
- Robotics and automation  
- Stock trading bots  

**Common Algorithms:**  
- Q-Learning  
- Deep Q-Networks (DQN)  
- Policy Gradient Methods  
- SARSA  

üß© *Example:* An AI learns to play chess by trying different moves and getting rewarded when it wins or punished when it loses.

### âœ¨ Intuition Behind KNN

KNN assumes:

> **Similar things exist near each other.**  
> Points that are close in feature space will likely have similar output values.

To predict output for a new point:
1. Compute distance to all training points  
2. Pick **K nearest neighbors**  
3.  
   - **Classification** â†’ output = majority class among K neighbors  
   - **Regression** â†’ output = average of neighbor values

### ðŸŽ¯ When to use KNN?
Use KNN when:
- Dataset is small-to-medium  
- Features are numeric  
- Decision boundary is nonlinear  
- You want simple, interpretable models  

Avoid KNN when:
- Data is huge (slow, distance computation expensive)  
- High dimensional dataset (curse of dimensionality)

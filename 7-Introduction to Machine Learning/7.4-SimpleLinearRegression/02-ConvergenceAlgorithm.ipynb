### What Are Convergence Algorithms?

A convergence algorithm is an optimization process that iteratively updates parameters (like weights in ML models) until it reaches a point where further updates make little or no improvement in the objective (usually the cost function).

In short:

Convergence = when the algorithm has learned enough ‚Äî the cost function has stopped changing significantly.

‚öôÔ∏è Formal Definition

An algorithm is said to converge if the sequence of its solutions approaches a stable value (usually the minimum or maximum of a function).

Mathematically:
	limit k -> infinity (Œ∏k‚Äã=Œ∏‚àó)

where

- ùúÉùëò : parameter at iteration 

- ùúÉ‚àó : optimal parameter (minimum of cost function)

```python
import numpy as np
import matplotlib.pyplot as plt

# Cost function: J = (x-3)^2 + 2
def J(x):
    return (x - 3)**2 + 2

# Gradient: dJ/dx = 2*(x-3)
def dJ(x):
    return 2 * (x - 3)

# Gradient Descent Loop
x = 10     # Start far from minimum
alpha = 0.2  # Learning rate
points = [x]
for _ in range(15):
    x = x - alpha * dJ(x)
    points.append(x)

# Plot
x_vals = np.linspace(0, 6, 100)
plt.plot(x_vals, J(x_vals), label="Cost Function")
plt.scatter(points, [J(p) for p in points], color="red")
plt.title("Convergence of Gradient Descent")
plt.xlabel("x (parameter)")
plt.ylabel("J(x) (cost)")
plt.legend()
plt.show()
```

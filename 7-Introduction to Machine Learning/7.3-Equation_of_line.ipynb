## Equation of a Line (2D)

A line in 2D space can be represented in multiple forms:

1. **Slope-Intercept Form:**
  
   y = mx + c
  
   where  
   - m  = slope of the line  
   - c = y-intercept (the point where the line cuts the y-axis)

2. **Point-Slope Form:**
   
   y - y_1 = m(x - x_1)

   where  
   - (x_1, y_1)  is a known point on the line.

3. **General Form:**
   
   Ax + By + C = 0



```python
import matplotlib.pyplot as plt
import numpy as np

# Equation: y = 2x + 1
x = np.linspace(-5, 5, 100)
y = 2 * x + 1

plt.figure(figsize=(6, 4))
plt.plot(x, y, label="y = 2x + 1")
plt.title("Equation of Line (2D)")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

## Equation of a Line in 3D

In 3D space, a line can be represented in **vector** or **parametric** form.

**Vector Form:**

vec{r} = vec{a} + lambda vec{b}


where:
- vec{a} = (x_1, y_1, z_1)  is a point on the line  
- vec{b} = (l, m, n) is the direction vector  
- lambda is a scalar parameter

**Parametric Form:**

x = x_1 + l\lambda

quad y = y_1 + m lambda

 quad z = z_1 + n lambda




```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')

# Line passing through (1,2,3) with direction vector (2, -1, 1)
t = np.linspace(-5, 5, 100)
x = 1 + 2 * t
y = 2 - 1 * t
z = 3 + 1 * t

ax.plot3D(x, y, z, 'b', label="Line in 3D")
ax.scatter(1, 2, 3, color='r', s=50, label="Point (1,2,3)")
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.title("Equation of a Line in 3D")
plt.show()
```

## Equation of a Plane (Hyperplane in 3D)

A plane in 3D (also called a hyperplane when generalized to higher dimensions) is represented as:


Ax + By + Cz + D = 0


where  
- \( A, B, C \) = normal vector components (perpendicular to the plane)  
- \( D \) = constant term

**Vector Form:**

\vec{r} \cdot \vec{n} = d

where  
- vec{n} = normal vector to the plane  
- d = perpendicular distance from origin


```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')

# Plane: x + y + z = 5
x = np.linspace(0, 5, 10)
y = np.linspace(0, 5, 10)
X, Y = np.meshgrid(x, y)
Z = 5 - X - Y

ax.plot_surface(X, Y, Z, alpha=0.6, color='cyan')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title("Plane: x + y + z = 5")
plt.show()
```

## Hyperplane in Higher Dimensions

In general, a **hyperplane** in *n-dimensional space* is represented as:


W1_X1 + W2_X2 + ... + Wn_Xn + b = 0


where  
- Wi are weights (coefficients)  
- Xi  are features  
- b is bias or intercept  

In Machine Learning, hyperplanes are used by algorithms like **Support Vector Machines (SVMs)** to **separate data points into classes**.


```python
from sklearn import datasets
from sklearn.svm import SVC
import matplotlib.pyplot as plt

# Load data
X, y = datasets.make_blobs(n_samples=100, centers=2, random_state=6)

# Fit a Support Vector Classifier
clf = SVC(kernel='linear')
clf.fit(X, y)

# Plot data and decision boundary
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr')
ax = plt.gca()

# Plot decision boundary (hyperplane)
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx = np.linspace(xlim[0], xlim[1])
yy = -(clf.coef_[0][0] * xx + clf.intercept_[0]) / clf.coef_[0][1]

plt.plot(xx, yy, 'k--', label='Hyperplane')
plt.legend()
plt.title("Hyperplane separating two classes (SVM)")
plt.show()
```

# **Eigen Decomposition on Covariance Matrix (PCA Intuition)**

Eigen decomposition is the mathematical core of **Principal Component Analysis (PCA)**.
To understand PCA, you must understand how eigenvalues and eigenvectors are obtained from the **covariance matrix**.

---

###  **1. What is a Covariance Matrix?**

Covariance matrix shows how features vary **together**.



---

###  **2. Why Perform Eigen Decomposition?**

Because eigenvalues and eigenvectors give us:

**Eigenvectors  Directions of Maximum Variance**

These are the **Principal Components (PCs)**.

**Eigenvalues  Amount of Variance Captured**

Larger eigenvalue = more important direction.

### Summary (Easy to Remember)

| PCA Step          | Meaning                             |
| ----------------- | ----------------------------------- |
| Covariance Matrix | Shows how variables vary together   |
| Eigenvectors      | Direction of maximum variance  PCs |
| Eigenvalues       | Amount of variance captured         |
| Choose top k      | Reduce dimensions                   |
| Project data      | Convert to new low-dim space        |

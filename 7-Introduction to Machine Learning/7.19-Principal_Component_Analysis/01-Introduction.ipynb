# **What is Dimensionality Reduction?**

**Dimensionality Reduction** is the process of reducing the number of input features (dimensions) while keeping the most important information.

### **Why do we need it?**

Because real-world datasets can have:

* Too many features
* Highly correlated features
* Noisy or redundant data
* Hard-to-visualize high dimensions

Dimensionality reduction helps in:

* Prevent from Curse of Dimensionality.
* Reducing overfitting
* Making models faster
* Improving accuracy
* Visualizing high-dimensional data (2D/3D plots)
* Removing noise

---

### **Types of Dimensionality Reduction**

**1. Feature Selection**

Choose some features and drop others.
Examples:

* Filter methods (correlation, chi-square)
* Wrapper methods (RFE)
* Embedded methods (Lasso)

**2. Feature Extraction (Transforms features)**

Create new features by combining old features.
Examples:

* PCA
* t-SNE
* UMAP
* Autoencoders

---

# **What is PCA? (Principal Component Analysis)**

PCA is the **most popular** dimensionality reduction algorithm.

It transforms the original high-dimensional features into **new features** called **Principal Components**.

These components:

* Are **linear combinations** of original features
* Are **uncorrelated**
* Capture **maximum variance** (information) in the data
* PC1 captures the most variance
* PC2 captures the second most, and so on

---

### **Simple Intuition Behind PCA**

Imagine you have a dataset with many features.
Some features might be:

* Highly correlated
* Redundant
* Not important

PCA rotates the coordinate system in such a way that:

* The **new axes (principal components)** align with **directions of maximum variance**
* It keeps the **important information**
* It drops less important dimensions

This reduces dimensions while preserving the structure.

---

### **Steps of PCA (Easy to Understand)**

1. **Standardize the data**
   (important because features may be in different scales)

2. **Compute the covariance matrix**

3. **Find eigenvalues & eigenvectors**

   * Eigenvectors  principal components
   * Eigenvalues  amount of variance captured

4. **Sort components by variance**

5. **Select top k components**
   (like PC1, PC2  top 2 components)

6. **Transform the original data**

---

### **Example**

Suppose your dataset has **50 features**.

Using PCA, you might reduce it to:

* **10 principal components**
* With **98% of the original information retained**

Faster, simpler, and less noise.

---

### **When to Use PCA?**

Use PCA when:

* Dataset has many features
* Features are correlated
* You want faster training
* You need to visualize high-dimensional data
* You want to remove noise

---

### **Visualization Example**

PCA can convert 50-dimensional data  2 dimensions
Then you can plot it on a 2D graph.



---

### **Key Advantages of PCA**

* Removes correlated features
* Makes models faster
* Helps visualization
* Reduces noise
* Improves accuracy (sometimes)

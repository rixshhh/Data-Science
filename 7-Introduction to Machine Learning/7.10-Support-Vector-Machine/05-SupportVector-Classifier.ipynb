```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Lets create a synthetic data points

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

X,y = make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

### Train , Test and Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

### Support Vector Classifier

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report , confusion_matrix

linear = SVC(kernel='linear')

linear.fit(X_train,y_train)
y_pred_linear = linear.predict(X_test)

cm = confusion_matrix(y_test, y_pred_linear)

print('Confusion-Matrix : ')
print(cm)

print('\nClassification_report')
print(classification_report(y_test,y_pred_linear))
```

```python
rbf = SVC(kernel='rbf')

rbf.fit(X_train,y_train)
y_pred_rbf = rbf.predict(X_test)

cm = confusion_matrix(y_test, y_pred_rbf)

print('Confusion-Matrix : ')
print(cm)

print('\nClassification_report')
print(classification_report(y_test,y_pred_rbf))
```

```python
poly = SVC(kernel='poly')

poly.fit(X_train,y_train)
y_pred_poly = poly.predict(X_test)

print('Confusion-Matrix : ')
print(confusion_matrix(y_test,y_pred_poly))

print('\nClassification_report')
print(classification_report(y_test,y_pred_poly))
```

```python
sigmoid = SVC(kernel='sigmoid')

sigmoid.fit(X_train,y_train)
y_pred_sigmoid = sigmoid.predict(X_test)

print('Confusion-Matrix : ')
print(confusion_matrix(y_test,y_pred_sigmoid))

print('\nClassification_report')
print(classification_report(y_test,y_pred_sigmoid))
```

### HyperParameter Tuning

```python
from sklearn.model_selection import GridSearchCV

params = {
    # Regularization parameter
    'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],

    # Kernel type to be tested
    'kernel': ['rbf'],

    # Kernel coefficient (used in rbf, poly and sigmoid)
    'gamma': ['scale', 'auto', 1, 0.1, 0.01, 0.001, 0.0001],

    # Whether to use shrinking heuristic
    'shrinking': [True, False],

    # Decision function shape
    'decision_function_shape': ['ovo', 'ovr']
}
```

```python
GridSearchCV(SVC(), param_grid=params , cv=5 , verbose = 3)
```

```python
grid = GridSearchCV(SVC(), param_grid=params , cv=5 , verbose = 3 , refit=True)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

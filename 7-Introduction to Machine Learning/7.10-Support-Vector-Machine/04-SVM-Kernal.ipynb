## SVM Kernel — Introduction

## Definition

A **Kernel** in Support Vector Machine (SVM) is a **mathematical function** that allows the algorithm to **transform data into a higher-dimensional space**,
so that it becomes **linearly separable** even if it was **non-linear in the original space**.

In Short :  

    If data cannot be seperated by the straight line , SVM usea Kernal function to transform data into a higher dimensions where it becomes seperable.

    - Imagine , points in a circle (Not seperable in 2D)

    - With a Kernal , Map data to 3D , Now we can seperate it with a Plane.


## Intuition

Imagine data that can’t be separated by a straight line in 2D.
A kernel maps it into a higher dimension (like 3D) where it can be separated by a plane.

## For example:

- In 2D → classes look like circles (non-linear)

- In 3D → they can be separated by a plane

## Mathematical Idea

Instead of explicitly transforming data with a function ϕ(x),
SVM uses a Kernel function K(x₁, x₂) that computes:

So we never actually calculate ϕ(x) —

this is called the **“Kernel Trick”** (saves computation time).

### When to use Which Kernal

| Kernel         | When to Use                                                      |
| -------------- | ---------------------------------------------------------------- |
| **Linear**     | Data is linearly separable or has many features (like text data) |
| **Polynomial** | Data has curved relationships                                    |
| **RBF**        | Default choice for most non-linear problems                      |
| **Sigmoid**    | Rarely used; similar to neural network activation                |


```python
from sklearn.svm import SVC
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Non-linear dataset
X, y = make_moons(n_samples=100, noise=0.1, random_state=42)

# Feature scaling
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

#### Linear Kernel SVM

```python
# Linear Kernel SVM
model_linear = SVC(kernel='linear')
model_linear.fit(X, y)

# Plot
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm')
plt.title("🔹 SVM with Linear Kernel")
plt.show()
```

#### Polynomial Kernel SVM

```python
# Polynomial Kernel SVM
model_poly = SVC(kernel='poly', degree=3)
model_poly.fit(X, y)

# Plot
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm')
plt.title("🔸 SVM with Polynomial Kernel (degree=3)")
plt.show()
```

#### RBF Kernel SVM

```python
# RBF Kernel SVM
model_rbf = SVC(kernel='rbf', gamma=1)
model_rbf.fit(X, y)

# Plot
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm')
plt.title("🧠 SVM with RBF Kernel")
plt.show()
```

#### Sigmoid Kernel SVM

```python
# Sigmoid Kernel SVM
model_sigmoid = SVC(kernel='sigmoid')
model_sigmoid.fit(X, y)

# Plot
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm')
plt.title("⚙️ SVM with Sigmoid Kernel")
plt.show()
```

### 2. Bayes‚Äô Theorem

Bayes‚Äô Theorem provides a way to update the probability estimate for a hypothesis
as more evidence or information becomes available.

The formula is:

    P(A|B) = (P(B|A) * P(A)) / P(B)

Where:
- P(A|B): Posterior probability ‚Äî probability of A given B
- P(B|A): Likelihood ‚Äî probability of B given A
- P(A): Prior probability ‚Äî probability of A before seeing data
- P(B): Evidence ‚Äî total probability of B

Suppose we want to predict if an email is 'Spam' or 'Not Spam' based on the presence of certain words.

We calculate:

    P(Spam | Words) = (P(Words | Spam) * P(Spam)) / P(Words)

We then compare it with:

    P(NotSpam | Words)

Whichever probability is higher ‚Äî that's our prediction.


### 3. Types of Na√Øve Bayes Classifiers


1. Gaussian Na√Øve Bayes:
   - Used when features are continuous and follow a normal (Gaussian) distribution.
   - Example: Iris dataset (flower measurements)

2. Multinomial Na√Øve Bayes:
   - Used for discrete data, like word counts in text classification.
   - Example: Spam detection, sentiment analysis.

3. Bernoulli Na√Øve Bayes:
   - Used when features are binary (0 or 1).
   - Example: Whether a word is present in an email or not.

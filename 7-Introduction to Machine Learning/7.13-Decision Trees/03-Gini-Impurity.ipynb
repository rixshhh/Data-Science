# Decision Tree - Gini Impurity



Gini Impurity is a measure used in Decision Trees (CART algorithm) to decide how good a split is.

It tells how mixed the classes are in a dataset.

## Definition

Gini impurity measures the probability that a randomly chosen sample would be incorrectly classified if it was labeled according to the class distribution in the node.

## Intuition Behind Gini

- If a node contains only one class,
     Gini = 0 (perfectly pure)

- If a node contains mixed classes,
     Gini increases

Maximum Gini occurs when classes are equally mixed

Example: 50% Class A, 50% Class B

## Why Decision Trees Use Gini

- Gini Impurity is fast to compute.

- Leads to good splits in practice.

- Works well for classification tasks.

- Preferred in CART (Classification and Regression Trees).

## Summary

| Concept       | Meaning                                             |
| ------------- | --------------------------------------------------- |
| Gini Impurity | Measures impurity (mixing of classes)               |
| Range         | 0  0.5 (for 2 classes), up to 1 (for many classes) |
| Best Split    | One that **reduces Gini impurity the most**         |
| Pure Node     | Gini = 0                                            |
| Most Impure   | When classes are equally distributed                |


## Gini Impurity vs Entropy (Information Gain)

Both Gini and Entropy are metrics used in Decision Trees to decide the best split.

They both measure impurity, but they use different formulas and have slightly different behaviors.

## Formula




## Intuition

**Gini**

- Measures how often a sample is incorrectly labeled

- Easier & faster to compute

- Prefers the largest class (more biased)

**Entropy**

- Measures the amount of disorder / randomness

- Uses logarithms  heavier computation

- More sensitive to class imbalance

## Range

| Metric      | Range            | Interpretation               |
| ----------- | ---------------- | ---------------------------- |
| **Gini**    | 0  0.5 (binary) | 0 = pure, 0.5 = max impurity |
| **Entropy** | 0  1            | 0 = pure, 1 = max impurity   |


## When do they give different results?

 Often they choose the same best split

In many practical datasets, both metrics identify similar or identical splits.

 Differences appear when:

- Classes are very imbalanced.

- There are multiple categories.

- You want a split that emphasizes information content.


## Practical Differences

| Feature           | Gini Impurity                   | Entropy (Information Gain) |
| ----------------- | ------------------------------- | -------------------------- |
| Speed             | Faster (no logs)                | Slower (log calculations)  |
| Tree Algorithm    | CART                            | ID3, C4.5, C5.0            |
| Sensitivity       | Less sensitive to changes       | More sensitive             |
| Bias              | Biased towards dominant class   | Unbiased                   |
| Decision Boundary | Similar performance in practice | Similar performance        |


## Which one should you use?

 Use Gini when:

- You want faster computation

- Using CART (sklearn DecisionTreeClassifier uses Gini by default)

 Use Entropy when:

- You want more information-theoretic interpretation

- You care about class imbalance sensitivity

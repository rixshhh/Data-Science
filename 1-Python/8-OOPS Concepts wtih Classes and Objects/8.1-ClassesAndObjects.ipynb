## Instance Variable and Methods

### Instance Variable

```python
class Dog:
    ## Constructor
    def __init__(self,name,age):
        ## Instance variable
        self.name=name
        self.age=age

### Creating Objects
dog1=Dog()
```

```python
dog1=Dog('dinno',4)
print(dog1)
print(dog1.name)
print(dog1.age)
```

```python
dog2=Dog('luccy',10)
print(dog2)
print(dog2.name)
print(dog2.age)
```

### Instance methods (Functionality)

```python
### Define a class with instance methods
class Dog():
    def __init__(self,name,age):
        self.name=name
        self.age=age

    ## instance methods
    def bark(self):
        print(f'{self.name} says woof')

dog1=Dog('dinno',4)
dog1.bark()
dog2=Dog('luccy',10)
dog2.bark()
```

## Modeling a Bank Account

```python
### Define a class for bank aaccount

class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance
    
    def deposit(self,amount):
        self.balance+=amount
        print(f'{amount} is deposited . New Balance is {self.balance}')

    def withdraw(self,amount):
        if amount>self.balance:
            print('Insufficient funds!')
        else:
            self.balance-=amount
            print(f"{amount} is withdrawn. New Balance is {self.balance}")

    def get_balance(self):
        return self.balance

### Creating an Object
account=BankAccount('Rishi',1000)
print(account.owner)
print(account.balance)
```

```python
## Calling instance methods
account.deposit(1000)
```

```python
account.withdraw(500)
```

```python
print(account.get_balance())
```

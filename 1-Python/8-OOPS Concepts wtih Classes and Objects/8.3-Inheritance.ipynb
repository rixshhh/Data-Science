# Inheritance In Python

Inheritance is a fundamental concept in OOPS that allow a class to inherit the attributes and methods from another class. This will cover single inheritance and multiple inheritance , demonstrating how to create and use them in Python.

Inheritance basically derived from the real-world scenarios.

#### Single Inheritance

```python
### Inheritance

### Parent class - Blue print of a car object
class Car:
    def __init__(self,windows,doors,engineType):
        self.windows = windows
        self.doors = doors
        self.engineType = engineType

    
    def drive(self):
        print(f'The person will drive the {self.engineType} car')
```

```python
car1=Car(4,4,'Petrol')
car1.drive()
```

```python
### In order to make Tesla inherit the car

# class child_Class(Parent_Class)
class tesla(Car):
    def __init__(self,windows,doors,engineType,is_selfDriving):
        ### These three attribute should be getting from the car class
        super().__init__(windows,doors,engineType)
        self.is_selfDriving = is_selfDriving
    
    def selfDriving(self):
        print(f'Tesla supports selfDriving :  {self.is_selfDriving} ')
```

```python
tesla1=tesla(4,5,'electric',True)
tesla1.selfDriving()
```

```python
tesla1.drive()
```

#### Multiple Inheritance

When a class inherits from more than one base class.

```python
### Base Class
class Animal:
    def __init__(self,name):
        self.name=name
    
    def speak(self):
        print('Subclass must implement this method')

### Another Base Class
class Pet:
    def __init__(self,owner):
        self.owner = owner

### Derived Class
class Dog(Animal,Pet):
    ### Not using super()
    def __init__(self,name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)
    
    def speak(self):
        return f'{self.name} say woof'
    
### Create an Object
dog=Dog("Buddy","Rishi")
print(dog.speak())
print(f'Owner:{dog.owner}')
```

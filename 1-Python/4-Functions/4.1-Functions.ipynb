### Variable length Arguments
* Positional Arguments

```python
## Positional Arguments
def print_numbers(*args): # *args will be having multiple parameters
    for numbers in args:
        print(numbers)
```

```python
## This parameters are basically called as Positional Arguments
print_numbers(1,2,3,4,5,'Rishikesh',3.14)
```

* Keyword Arguments -> given by **kwargs

In keyword Arguments , All the parameter will be in the form of key value arguments like key-value pairs.

```python
def print_details(**kwargs): # **kargs will be having multiple parameters
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(Name='Rishikesh',Age='22',Country='India')
```

(Name='Rishikesh',Age='22',Country='India') 

All these parameter will be referenced by **kwargs



### Combine both of them

```python
def print_details(*args,**kwargs): 
    for val in args:
        print(f"Positional Arguments : {val}")

    for key,value in kwargs.items():
        print(f"\nKeyword Arguments : {key}:{value}")
```

```python
print_details(1,2,Name='Rishikesh',Age='22',Country='India' ,3,4,5,'Rishi')
```

```python
print_details(1,2,3,4,5,'Rishi',Name='Rishikesh',Age='22',Country='India')
```

### Return Statement

In return statement , our function return something right after the entire statement is executed

```python
def multiply(a,b):
    return (f"{a} * {b} = {a*b}")
multiply(2,3)
```

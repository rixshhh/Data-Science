# Iterators

Iterators are advanced Python concepts that allow for efficient looping and memory management. Iterators provide a way to access elements of a collection sequentially without exposing the underlying structure.

```python
My_list=[1,2,3,4,5,6,7]
for i in My_list:
    print(i)
```

```python
type(My_list)
```

```python
print(My_list)
```

```python
### Iterator
iterator=iter(My_list)
print(type(iterator))
```

```python
iterator
```

It uses lazy loading technique. That basically means my list_iterator is created in some memory location. It will not give us any elements unless and until we iterate through that particular element.

Now, in order to iterate through all the elements , we have to use another function which is called as next.

And here if we go ahead and write iterator as soon as, we probably go ahead and hit next of iterator, then what is the first element in this list that will get displayed.

```python
### In order to iterate through all the element
next(iterator)
```

```python
next(iterator)
```

```python
next(iterator)
```

```python
My_list=[1,2,3]
iterator=iter(My_list)
```

```python
try:
    print(next(iterator))
except StopIteration:
    print('There are no element in the iterator.')
```

### String Iterator

```python
my_string='Rishi'
string_iterator = iter(my_string)

try:
    print(next(string_iterator))
except StopIteration:
    print('There are no element in the iterator.')
```

```python
try:
    print(next(string_iterator))
except StopIteration:
    print('There are no element in the iterator.')
```

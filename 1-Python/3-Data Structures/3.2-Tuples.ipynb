### Immutable Nature of Tuple

Tuple are immutable, meaning their element cannot be changed once assigned.

```python
## List are Mutable
my_list = [1,2,3,4,5]
print(my_list)

my_list[2] = 'Replaced'
print(my_list)
```

```python
my_tuple = (1,2,3,4,5)
print(my_list)

my_tuple[2] = 'Replaced'
print(my_tuple)
```

That's why Tuples are Immutable

### Packing and UnPacking Tuples

```python
## Packing a tuple
packed_tuples = 1,'hello-world', 3.14
print(packed_tuples)
```

Now whenever we define the numbers in this way like comma separated , by default what it is going to do is that whatever number of elements that we've given, IT IS GOING TO PACK THEM IN A TUPLE ITSELF.

```python
## UnPacking a tuple
a,b,c =packed_tuples
print(a)
print(b)
print(c)
```

```python
## Unpacking with * (star)
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers

print(first)
print(middle)
print(last)
```

So what this star does is that from wherever you are probably starting till the last element,
right before that last element, all the elements will be picked up.

### Nested Tuple

tuples inside a tuple

```python
## Nested List
mylist = [[1,2,3],[4,5,6],[1,'hello',3.14]]
print(mylist[0])
mylist[0][0:2]
```

```python
nested_tuple = ((1, 2, 3), ("a", "b", "c"), (True, False))
```

### Access the elements inside a tuple

```python
print(nested_tuple[0])
print(nested_tuple[1][2])
```

### Iterating over nested tuples

```python
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```

## âš™ï¸ Basic Flask App
```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return 'Hello, Flask!'

if __name__ == '__main__':
    app.run(debug=True)
```

## ğŸ§© Routing
Routing is the process of mapping URLs to functions in Flask.

```python
@app.route('/about')
def about():
    return 'This is the About Page'
```

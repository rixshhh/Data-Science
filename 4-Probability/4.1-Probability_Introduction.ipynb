
# **Introduction to Probability**

**Probability** is the branch of mathematics that deals with measuring the likelihood of an event occurring.  
It helps us **quantify uncertainty** and make predictions about future outcomes.

In simple words, **probability tells us how likely something is to happen.**



## **Key Points**
- The value of probability always lies between **0 and 1**.  
  - **0** means the event is **impossible**.  
  - **1** means the event is **certain** to happen.  
- Probabilities can also be expressed as **fractions**, **decimals**, or **percentages**.



## **Example**
When you toss a fair coin:  
- Probability of getting a **Head (H)** = ½ = 0.5 = 50%  
- Probability of getting a **Tail (T)** = ½ = 0.5 = 50%



## **Formula for Probability**
Probability of an event (E):  
**P(E) = (Number of favorable outcomes) / (Total number of possible outcomes)**


```python
# Example 1: Probability of getting heads in a coin toss
favorable_outcomes = 1  # Head
total_outcomes = 2  # Head or Tail

P_E = favorable_outcomes / total_outcomes
print("Probability of getting a Head:", P_E)
```

```python
# Example 2: Probability of rolling an even number on a 6-sided dice
favorable_outcomes = 3  # {2, 4, 6}
total_outcomes = 6

P_E = favorable_outcomes / total_outcomes
print("Probability of rolling an even number:", P_E)
```


### **Conclusion**
Probability helps us understand the chances of events happening in real life — whether it’s predicting the weather, outcomes in games, or making business decisions based on data.

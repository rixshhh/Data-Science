
# **Addition Rule of Probability (Mutual and Non-Mutual Exclusive Events)**

The **Addition Rule** is used to find the probability that **at least one of two events** will occur.



## **1. Mutually Exclusive Events**

Two events are **mutually exclusive** (or disjoint) if they **cannot occur at the same time**.

### **Formula (in words):**
P(A ∪ B) = P(A) + P(B)

### **Example:**
When rolling a die:  
Let event A = getting a 2, and event B = getting a 5.  
Since a single roll cannot show both 2 and 5 at the same time, these are **mutually exclusive**.

P(A) = 1/6, P(B) = 1/6  
P(A ∪ B) = 1/6 + 1/6 = 2/6 = **0.333**


```python

# Example 1: Mutually Exclusive Events
P_A = 1/6  # Probability of getting 2
P_B = 1/6  # Probability of getting 5

P_A_or_B = P_A + P_B
print("Probability of getting 2 or 5 (Mutually Exclusive):", P_A_or_B)
```


## **2. Non-Mutually Exclusive Events**

Two events are **non-mutually exclusive** if they **can occur together**.

### **Formula (in words):**
P(A ∪ B) = P(A) + P(B) − P(A ∩ B)

### **Example:**
When drawing a card from a deck:  
Let event A = drawing a red card, and event B = drawing a King.  
Since there are red Kings (hearts, diamonds), these events **can occur together**.

P(A) = 26/52, P(B) = 4/52, P(A ∩ B) = 2/52  
P(A ∪ B) = 26/52 + 4/52 − 2/52 = 28/52 = **0.538**


```python
# Example 2: Non-Mutually Exclusive Events
P_A = 26/52  # Probability of red card
P_B = 4/52   # Probability of King
P_A_and_B = 2/52  # Probability of red King

P_A_or_B = P_A + P_B - P_A_and_B
print("Probability of drawing a red card or a King (Non-Mutually Exclusive):", P_A_or_B)
```


### **Conclusion**
- Use **P(A ∪ B) = P(A) + P(B)** when events are **mutually exclusive**.  
- Use **P(A ∪ B) = P(A) + P(B) − P(A ∩ B)** when events are **non-mutually exclusive**.  

Understanding this helps calculate combined probabilities correctly in real-world problems.

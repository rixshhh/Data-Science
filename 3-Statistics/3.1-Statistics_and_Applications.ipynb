# **Statistics and Its Applications in Daily Life**

## **Definition**
Statistics is the branch of mathematics that deals with the **collection, organization, analysis, interpretation, and presentation of data**.  
It helps us to make decisions and draw conclusions based on data.

---

## **Types of Statistics**
1. **Descriptive Statistics:**  
   It summarizes and describes the main features of a data set.  
   *Example:* Calculating the average marks of students in a class.

2. **Inferential Statistics:**  
   It involves drawing conclusions or making predictions about a population based on a sample of data.  
   *Example:* Predicting election results by surveying a group of voters.

---

## **Applications of Statistics in Daily Life**

1. **Education:**  
   Used to analyze studentsâ€™ performance and improve teaching methods.

2. **Health:**  
   Helps in studying diseases, testing new medicines, and predicting health trends.

3. **Business and Economics:**  
   Used for market research, sales forecasting, and understanding customer preferences.

4. **Government:**  
   Used in census, unemployment rate calculation, and policy-making.

5. **Sports:**  
   Used to evaluate player performance, team strategies, and game statistics.

6. **Weather Forecasting:**  
   Used to predict weather conditions using past data and patterns.

7. **Banking and Insurance:**  
   Used to assess risks, calculate premiums, and manage investments.

---

## **Conclusion**
Statistics plays a vital role in our daily life by helping us **understand data, make better decisions, and solve real-world problems** effectively.


```python
# Example: Simple Statistics using Python

import numpy as np
import pandas as pd

# Sample data: marks of 5 students
marks = [75, 80, 65, 90, 85]

# Calculating Mean, Median, and Standard Deviation
mean = np.mean(marks)
median = np.median(marks)
std_dev = np.std(marks)

print(f"Marks: {marks}")
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std_dev:.2f}")

# Creating a small DataFrame
df = pd.DataFrame({'Marks': marks})
df.describe()
```

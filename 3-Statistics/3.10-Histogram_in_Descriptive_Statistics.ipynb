# **Histogram in Descriptive Statistics**

A **histogram** is a graphical representation that organizes a group of data points into user-specified ranges (called *bins*). It helps visualize the **distribution** of a dataset.

## **1. What is a Histogram?**
A histogram shows the **frequency** (or count) of data values within certain intervals.

It is used to understand:
- The **shape** of the data distribution
- The **central tendency** (where values cluster)
- The **spread** or **variability**

### **Example:**
Let’s consider students' exam marks ranging from 0 to 100.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data (students' marks)
np.random.seed(42)
marks = np.random.normal(loc=70, scale=10, size=100)

plt.hist(marks, bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Students' Marks")
plt.xlabel("Marks Range")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

## **2. Understanding the Histogram**

- Each bar represents the **frequency** of marks within a range.
- The **height** of each bar shows how many students scored in that range.
- The **width** of bars is defined by the number of bins.

## **3. Using Pandas for Histogram**
You can also use the **pandas** library to create a histogram easily.

```python
import pandas as pd

# Create a DataFrame
data = pd.DataFrame({'Marks': marks})

# Plot using pandas
data['Marks'].plot(kind='hist', bins=10, color='lightgreen', edgecolor='black', title='Histogram using Pandas')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

## **4. Interpreting Histogram Shapes**

### **a) Symmetrical (Normal Distribution)**
- Data is evenly distributed around the mean.
- Example: Human height.

### **b) Positively Skewed (Right-Skewed)**
- Tail is longer on the right.
- Example: Income distribution.

### **c) Negatively Skewed (Left-Skewed)**
- Tail is longer on the left.
- Example: Age at retirement.

```python
# Plotting different histogram shapes

# Normal Distribution
normal_data = np.random.normal(50, 10, 1000)

# Right Skewed (Exponential)
right_skewed = np.random.exponential(10, 1000)

# Left Skewed (Reverse exponential)
left_skewed = -np.random.exponential(10, 1000) + 50

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes[0].hist(normal_data, bins=15, color='skyblue', edgecolor='black')
axes[0].set_title("Normal Distribution")

axes[1].hist(right_skewed, bins=15, color='lightcoral', edgecolor='black')
axes[1].set_title("Right-Skewed Distribution")

axes[2].hist(left_skewed, bins=15, color='lightgreen', edgecolor='black')
axes[2].set_title("Left-Skewed Distribution")

for ax in axes:
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## **5. Summary**
✅ A **Histogram** is used to visualize the frequency distribution of data.
✅ It helps understand the **shape**, **central tendency**, and **spread**.
✅ You can create histograms using **Matplotlib** or **Pandas**.
✅ Different shapes (normal, right-skewed, left-skewed) provide insights into data behavior.

# **Random Variables in Statistics**

A **random variable** is a variable whose possible values are the outcomes of a random phenomenon.

It assigns a **numerical value** to each outcome in a probability experiment.

## **1. Introduction**
In statistics, a random variable represents **uncertainty**. It can take different values based on the outcome of a random event.

### **Examples:**
- Rolling a die ðŸŽ² â†’ Outcomes: {1, 2, 3, 4, 5, 6}
- Tossing a coin ðŸª™ â†’ Outcomes: {Head = 1, Tail = 0}
- Measuring height of people â†’ continuous outcomes

## **2. Types of Random Variables:**

### **a) Discrete Random Variable**
Takes **countable** values.

**Examples:**
- Number of heads in 3 coin tosses
- Number of students present in a class

### **b) Continuous Random Variable**
Takes **uncountable (infinite)** values within a range.

**Examples:**
- Height of students
- Time taken to complete a task

## **3. Probability Distribution**
A **probability distribution** gives the probability of each possible value of a random variable.

For a discrete random variable `X`,
\[ P(X = x_i) = p_i \]
where \( \sum p_i = 1 \)

```python
import numpy as np
import matplotlib.pyplot as plt

# Example: Rolling a die
x = np.array([1, 2, 3, 4, 5, 6])
p = np.ones(6) / 6  # Equal probability for each outcome

plt.bar(x, p, color='skyblue', edgecolor='black')
plt.title("Probability Distribution of Rolling a Die")
plt.xlabel("Outcome (X)")
plt.ylabel("Probability P(X)")
plt.show()
```

## **4. Expected Value (Mean) and Variance**

### **Formulas:**
- **Expected Value (Mean):**  
  \( E(X) = \sum [x_i * P(x_i)] \)

- **Variance:**  
  \( Var(X) = \sum [(x_i - E(X))^2 * P(x_i)] \)

### **Example:** Rolling a fair die

```python
# Expected value and variance for rolling a die
E_X = np.sum(x * p)
Var_X = np.sum(((x - E_X) ** 2) * p)

print("Expected Value (Mean):", E_X)
print("Variance:", Var_X)
```

## **5. Continuous Random Variable Example (Normal Distribution):**

A continuous random variable is described by a **probability density function (PDF)**.

Hereâ€™s an example of a **normal distribution**:

```python
from scipy.stats import norm

# Define normal distribution
mu, sigma = 0, 1
x_vals = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x_vals, mu, sigma)

plt.plot(x_vals, pdf, color='purple')
plt.title("Normal Distribution (Î¼ = 0, Ïƒ = 1)")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

## **6. Summary:**
*  A **Random Variable** assigns numerical values to random outcomes.
* Two types: **Discrete** and **Continuous**.
* Discrete variables use **probability mass functions (PMF)**.
* Continuous variables use **probability density functions (PDF)**.
* Key measures: **Expected Value** and **Variance** help describe their behavior.

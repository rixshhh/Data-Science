# **Variables in Python**

A **variable** is a name that stores a value in memory so that it can be used and changed later in a program.

## **1. What is a Variable?**

A variable acts like a **container** that holds some data or information.

For example, if you write:
```python
x = 10
```
This means that a variable named `x` stores the value `10`.

## **2. Example:**

````
# Assigning values to variables
name = "Rishikesh"
age = 22
height = 5.9

# Displaying the variables
print("Name:", name)
print("Age:", age)
print("Height:", height)
````

## **3. Key Points about Variables:**
- A variable must have a **name** and a **value**.
- The value stored in a variable can be **changed** during program execution.
- Variables make code more **readable** and **dynamic**.

## **4. Rules for Naming Variables:**
✅ Must start with a **letter** or **underscore (_)**  
✅ Can contain **letters**, **numbers**, and **underscores**  
❌ Cannot start with a **number**  
❌ Cannot contain **spaces or special characters**  

### **Examples:**

````
# Valid variable names
_valid_name = "hello"
name1 = "John"

# Invalid variable names (will cause errors if uncommented)
# 1name = "error"
# name@ = "error" 
````

## **5. Types of Variables (Based on Scope):**

### **a) Local Variables:**
Declared inside a function and can only be used within that function.

### **b) Global Variables:**
Declared outside all functions and can be accessed anywhere in the program.

````
# Example of Local and Global Variables

x = 50  # Global variable

def show():
    y = 10  # Local variable
    print("Inside function, local variable y =", y)
    print("Inside function, global variable x =", x)

show()
print("Outside function, global variable x =", x)
````

## **6. Summary:**
- Variables are **names that store data**.
- They can hold different data types (int, float, string, etc.).
- You can **change** the value of a variable anytime.
- Follow proper **naming rules** to avoid errors.
- Understand **local** and **global** variable usage for clear code structure.

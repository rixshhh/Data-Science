# **5-Number Summary - Descriptive Statistics**

The **5-number summary** provides a quick overview of the distribution of a dataset. It includes five key values that describe the center and spread of data.

## **1. Components of the 5-Number Summary**
The five numbers are:

1. **Minimum (Min):** The smallest value in the dataset.
2. **First Quartile (Q1):** The 25th percentile — 25% of data lies below this value.
3. **Median (Q2):** The 50th percentile — the middle value.
4. **Third Quartile (Q3):** The 75th percentile — 75% of data lies below this value.
5. **Maximum (Max):** The largest value in the dataset.

Together, they give a **summary of data distribution and variability**.

```python
import numpy as np
import pandas as pd

# Generate random student marks data
np.random.seed(42)
marks = np.random.randint(30, 100, 50)
data = pd.DataFrame({'Marks': marks})

# Display first few values
data.head()
```

## **2. Calculating the 5-Number Summary using NumPy**

```python
# Calculate 5-number summary
minimum = np.min(marks)
q1 = np.percentile(marks, 25)
median = np.median(marks)
q3 = np.percentile(marks, 75)
maximum = np.max(marks)

print("Minimum:", minimum)
print("Q1 (25th Percentile):", q1)
print("Median (Q2):", median)
print("Q3 (75th Percentile):", q3)
print("Maximum:", maximum)

# Create summary dictionary
summary = {
    'Minimum': minimum,
    'Q1 (25th)': q1,
    'Median (Q2)': median,
    'Q3 (75th)': q3,
    'Maximum': maximum
}

summary
```

## **3. Visualizing the 5-Number Summary using a Boxplot**
A **boxplot** (box-and-whisker plot) visually represents the 5-number summary. It shows:
- **Box edges:** Q1 and Q3
- **Line inside the box:** Median (Q2)
- **Whiskers:** Min and Max values
- **Outliers:** Individual points beyond whiskers

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
plt.boxplot(marks, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Boxplot Representing 5-Number Summary')
plt.ylabel('Marks')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

## **4. Interpreting the Boxplot**

- The **bottom whisker** shows the **minimum**.
- The **bottom edge** of the box is **Q1 (25th percentile)**.
- The **line inside the box** is the **median (Q2)**.
- The **top edge** of the box is **Q3 (75th percentile)**.
- The **top whisker** represents the **maximum**.
- Any dots beyond whiskers indicate **outliers**.

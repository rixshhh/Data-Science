# **Correlation and Covariance - Descriptive Statistics**

Both **correlation** and **covariance** measure the **relationship between two variables** — how one variable changes when the other changes.

## **1. Covariance**

Covariance measures the **direction of the linear relationship** between two variables.

### **Formula:**
Cov(X, Y) = [ Σ (Xi − X̄)(Yi − Ȳ) ] / (n − 1)

### **Interpretation:**
- **Positive covariance:** When one variable increases, the other tends to increase.
- **Negative covariance:** When one variable increases, the other tends to decrease.
- **Zero covariance:** No linear relationship.

## **2. Correlation**

Correlation measures both the **strength and direction** of a linear relationship between two variables.

### **Formula:**
r = [ Cov(X, Y) ] / (σX × σY)

where:
- σX = standard deviation of variable X

- σY = standard deviation of variable Y

### **Interpretation:**
| Correlation (r) | Relationship |
|-----------------|---------------|
| +1 | Perfect Positive |
| 0 | No Linear Relation |
| -1 | Perfect Negative |

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a realistic dataset: Study Hours vs Marks
np.random.seed(42)
study_hours = np.random.randint(1, 10, 15)
marks = 5 * study_hours + np.random.randint(-5, 10, 15)

data = pd.DataFrame({'Study_Hours': study_hours, 'Marks': marks})
data
```

## **3. Covariance using NumPy and Pandas**

```python
# Covariance matrix using NumPy
cov_matrix = np.cov(data['Study_Hours'], data['Marks'])
print("Covariance Matrix:\n", cov_matrix)

# Covariance using Pandas
cov_value = data.cov().iloc[0,1]
print(f"Covariance between Study Hours and Marks: {cov_value:.2f}")
```

## **4. Correlation using NumPy and Pandas**

```python
# Correlation using NumPy
corr_value = np.corrcoef(data['Study_Hours'], data['Marks'])[0,1]
print(f"Correlation Coefficient (r): {corr_value:.2f}")

# Correlation using Pandas
data_corr = data.corr()
print("\nCorrelation Matrix:\n", data_corr)
```

## **5. Visualization**

```python
# Scatter Plot
plt.figure(figsize=(6,4))
plt.scatter(data['Study_Hours'], data['Marks'], color='skyblue', edgecolor='black')
plt.title('Scatter Plot - Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Heatmap of Correlation
import seaborn as sns
plt.figure(figsize=(5,4))
sns.heatmap(data_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

## **6. Interpretation**
✅ **Covariance** tells us the **direction** of the relationship (positive or negative).

✅ **Correlation** tells us both **direction** and **strength** (ranging from -1 to +1).

✅ **r ≈ +1:** Strong positive relationship.

✅ **r ≈ -1:** Strong negative relationship.

✅ **r ≈ 0:** No linear relationship.

## **7. Summary**
| Concept | Measures | Range | Interpretation |
|----------|-----------|--------|----------------|
| **Covariance** | Direction only | -∞ to +∞ | Positive or Negative relationship |
| **Correlation** | Direction + Strength | -1 to +1 | Strong, Weak, or No relation |

# **Percentiles and Quartiles - Descriptive Statistics**

In descriptive statistics, **percentiles** and **quartiles** are measures that divide a dataset into equal parts to understand the **distribution** and **spread** of the data.

## **1. What is a Percentile?**
A **percentile** indicates the value **below which a given percentage of observations** in a dataset fall.

### **Formula:**
Percentile = n/p x 100
where:
- *n* = number of data points
- *p* = desired percentile (e.g., 25th, 50th, 90th)

### **Example:**
If a student’s score is at the 80th percentile, it means they scored **better than 80%** of all students.

## **2. What is a Quartile?**
Quartiles divide the dataset into **four equal parts**.

- **Q1 (1st Quartile)** → 25th percentile → 25% of data below this point
- **Q2 (2nd Quartile / Median)** → 50th percentile → middle of the data
- **Q3 (3rd Quartile)** → 75th percentile → 75% of data below this point

### **Interquartile Range (IQR):**
Measures the **spread of the middle 50%** of the data.

\[ IQR = Q3 - Q1 \]

```python
import numpy as np
import pandas as pd

# Generate random students' marks
np.random.seed(42)
marks = np.random.randint(30, 100, 50)
data = pd.DataFrame({'Marks': marks})

# Display first 10 marks
data.head(10)
```

## **3. Calculating Percentiles using NumPy**

```python
# Calculate 25th, 50th, and 75th percentiles
p25 = np.percentile(marks, 25)
p50 = np.percentile(marks, 50)
p75 = np.percentile(marks, 75)

print(f"25th Percentile (Q1): {p25}")
print(f"50th Percentile (Q2 / Median): {p50}")
print(f"75th Percentile (Q3): {p75}")

# Calculate IQR
IQR = p75 - p25
print(f"Interquartile Range (IQR): {IQR}")
```

## **4. Visualization using Boxplot**
A **boxplot** (box-and-whisker plot) visually shows the quartiles, median, and possible outliers.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
plt.boxplot(marks, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Boxplot of Students' Marks")
plt.ylabel("Marks")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

## **5. Interpreting the Boxplot**

- The **bottom edge** of the box = Q1 (25th percentile)
- The **line inside the box** = Q2 (Median)
- The **top edge** of the box = Q3 (75th percentile)
- The **IQR** = Q3 − Q1 → middle 50% of data
- Points outside the whiskers → **outliers** (very high or low values)

## **6. Summary**
* **Percentiles** help identify how a particular value compares to the entire dataset.
* **Quartiles** divide the dataset into four parts — Q1, Q2, and Q3.
* **IQR** helps measure variability and detect outliers.
*  **Boxplots** are useful visual tools to summarize data distribution.

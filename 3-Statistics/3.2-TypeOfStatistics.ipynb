# **Statistics**

Statistics is divided into two main branches:

---

## **1. Descriptive Statistics**
It consists of **organizing and summarizing data.**

### **Measures of Central Tendency**
1. Mean  
2. Median  
3. Mode  

### **Measures of Dispersion**
1. Variance  
2. Standard Deviation  

These measures help in describing and summarizing a given set of data.

---

## **2. Inferential Statistics**
It deals with **drawing conclusions or inferences** about a population based on a sample of data.

### **Process:**
- **Collect Data** from a sample.  
- Use experiments or tests (e.g., *z-test, t-test*).  
- Draw **conclusions or inferences** about the population data.

---

### **Example:**
Suppose there is a college **A** with 1000 students (population).  
We take a **sample** of 5 students with heights:

```
Sample = {180cm, 170cm, 162cm, 150cm, 160cm}
```

**Descriptive Statistics:**  
- Mean = 165 cm (Average height)  
- Median, Mode, Variance, etc., can also be calculated.

**Inferential Statistics:**  
- From this sample data, we can make **conclusions or inferences** about the population of 1000 students.

---

### **Summary**
| Type | Purpose | Example |
|------|----------|----------|
| **Descriptive** | Summarize or describe data | Mean, Median, Mode |
| **Inferential** | Make conclusions/inferences about a population | Using z-test, t-test |


````
# Example: Calculating Mean and Standard Deviation in Python

import numpy as np

# Sample heights (in cm)
sample = [180, 170, 162, 150, 160]

# Calculations
mean = np.mean(sample)
std_dev = np.std(sample)

print(f"Sample: {sample}")
print(f"Mean (Average Height): {mean:.2f} cm")
print(f"Standard Deviation: {std_dev:.2f} cm")
````

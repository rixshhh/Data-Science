## Handling Imbalanced Dataset

## what is Imbalanced Dataset?

An imbalanced dataset refers to a situation in machine learning or statistics where the number of samples in different classes (categories) is not evenly distributed.

⚖️ Definition:

A dataset is imbalanced when one class (category) has significantly more samples than the others.

1. Up Sampling
* Upsampling means increasing the number of samples in the minority class so that it matches (or becomes closer to) the number of samples in the majority class.

2. Down Sampling
* Downsampling means reducing the number of samples in the majority class so that it matches the minority class.

```python
import numpy as np
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(123)

# Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0,n_class_1
```

```python
## CREATE MY DATAFRAME WITH IMBALANCED DATASET
DataFrame1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

DataFrame2 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})
```

```python
df=pd.concat([DataFrame1,DataFrame2]).reset_index(drop=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['target'].value_counts()
```

## Upsampling

```python
## upsampling
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
from sklearn.utils import resample
df_minority_upsampled=resample(df_minority,replace=True, #Sample With replacement
         n_samples=len(df_majority),
         random_state=42 # any randonState value so that the seed is fixed.
        )
```

```python
df_upsampled=pd.concat([df_majority,df_minority_upsampled])
```

## Down Sampling

```python
import numpy as np
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(123)

# Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0

DataFrame1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

DataFrame2 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})

df = pd.concat([DataFrame2, DataFrame2]).reset_index(drop=True)

# Check the class distribution
print(df['target'].value_counts())
```

```python
df.head()
```

```python
## downsampling
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
from sklearn.utils import resample
df_majority_downsampled=resample(df_minority,replace=False,
         n_samples=len(df_minority),
         random_state=42
        )
```

```python
df_majority_downsampled.shape
```

```python
df_downsampled=pd.concat([df_minority,df_majority_downsampled])
```

```python
df_downsampled['target'].value_counts()
```
